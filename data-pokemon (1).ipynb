```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pokemon_data = pd.read_csv('C:\\Users\\utente\\Pokemons.csv')

mean_value = pokemon_data['total'].mean()
std_dev = pokemon_data['total'].std()
outliers = pokemon_data[(pokemon_data['hp'] > mean_value + 3 * std_dev) | (pokemon_data['hp'] < mean_value - 3 * std_dev)]
plt.scatter(pokemon_data.index, pokemon_data['total'], label='total')
plt.scatter(outliers.index, outliers['total'], color='red', label='Outliers')


plt.axhline(y=mean_value, color='green', linestyle='--', label='Media')
plt.axhline(y=mean_value + 3 * std_dev, color='orange', linestyle='--', label='Â±3 Deviazioni Standard')
plt.axhline(y=mean_value - 3 * std_dev, color='orange', linestyle='--')

plt.xlabel('Indice')
plt.title('Grafico che evidenzia')
plt.legend()
plt.show()
```

```python

plt.scatter(outliers.index, outliers['total'], color='blue', label='Outliers')
plt.axhline(y=mean_value, color='green', linestyle='-', label='Media')
plt.axhline(y=mean_value + 3 * std_dev, color='orange', linestyle='-', label='Â±3 Deviazioni Standard')
plt.axhline(y=mean_value - 3 * std_dev, color='orange', linestyle='-')

plt.xlabel('Indice')
plt.ylabel('punti totali')
plt.title('Outliers dei punti totali')
plt.legend()
plt.show()
```

```python
# ordinale scaler
import pandas as pd

pokemon_data = pd.read_csv('pokemons.csv')

tipi_pokemon = pokemon_data['type1'].unique()
tipi_pokemon = pokemon_data['type2'].unique()

ordine_tipi = {tipo: i+1 for i, tipo in enumerate(tipi_pokemon)}

data = {'Tipo': ['water', 'fire', 'grass']}
df = pd.DataFrame(data)

df['Tipo_Ordinale'] = df['Tipo'].map(ordine_tipi)

print(df)
```
